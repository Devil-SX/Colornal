```python
import numpy as np
from Colornal import Colornal
```

亮度渐变

```python
N = 100
x = np.linspace(0, 1, N)

h = 0.5*np.ones(N)
l = 0.5*np.sin(2*np.pi*x) + 0.5
s = 0.5*np.ones(N)
cs = Colornal(h,l,s)
cs.draw()
```

亮度渐变

```python
N = 100
x = np.linspace(0, 1, N)

h = 0.5*np.ones(N)
l = 0.5*np.sin(np.pi/2*x) + 0.5
s = 0.5*np.ones(N)
cs = Colornal(h,l,s)
cs.draw()
```

定饱和渐变

```python
N = 500
x = np.linspace(0, 1, N)

h = 0.5*np.sin(np.pi/2*x) + 0.5
l = 0.5*np.sin(np.pi/2*x) + 0.5
s = 0.5*np.ones(N)
cs = Colornal(h,l,s)
cs.draw()
```

```python
N = 500
x = np.linspace(0, 1, N)

h = 0.5*np.sin(np.pi*x-np.pi/2) + 0.5
l = 0.5*np.sin(np.pi/2*x) + 0.5
s = 0.5*np.ones(N)
cs = Colornal(h,l,s)
cs.draw()
```

```python
N = 500
x = np.linspace(0, 1, N)

h = 0.5*np.sin(np.pi/4*x+np.pi) + 0.5
l = 0.5*np.sin(np.pi/2*x) + 0.5
s = 0.5*np.ones(N)
cs = Colornal(h,l,s)
cs.draw()
```
